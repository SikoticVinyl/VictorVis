```python
# Drop the specified columns
player_stats_df.drop(columns=['URL', 'country'], inplace=True)

# Get the list of object columns
dataframe_objects = player_stats_df.select_dtypes(include='object').columns.tolist()

# Print the object columns
print(dataframe_objects)
```

```python
from fractions import Fraction

def convert_percentage(value):
    return float(value.replace('%', '')) / 100

def convert_fraction(value):
    return float(Fraction(value))

def convert_time_to_seconds(time_str):
    minutes, seconds = 0, 0
    if 'm' in time_str:
        minutes = int(time_str.split('m')[0].strip())
    if 's' in time_str:
        seconds = int(time_str.split('m')[-1].replace('s', '').strip())
    return minutes * 60 + seconds
```

```python
conversion_map = {
    'percent_col': convert_percentage,
    'fraction_col': convert_fraction,
    'time_col': convert_time_to_seconds,
}
```

```python
for column in dataframe_objects:
    if column in conversion_map:
        player_stats_df[column] = player_stats_df[column].apply(conversion_map[column])
```
